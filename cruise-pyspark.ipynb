```python
#Dataframe API
#Vector assembler is grouping features together 
# [Age, Tonnage] ----> new feature ---> independent feature
from pyspark.ml.feature import VectorAssembler
featureassembler=VectorAssembler(inputCols=['Age',
               'Tonnage',
               'passengers',
               'length',
               'cabins',
               'passenger_density',
               'crew'],outputCol="Independent feature")
```

```python
output=featureassembler.transform(df_cruise)
```

```python
output.show()
```

```python
finalized_data=output.select("Independent feature","passengers")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
train_data,test_data=finalized_data.randomSplit([0.75,0.25])
regressor=LinearRegression(featuresCol="Independent feature", labelCol="passengers")
regressor=regressor.fit(train_data)
```

```python
##Coefficients
regressor.coefficients
```

```python
#Intercepts
regressor.intercept
```

```python
#Prediction
pred_results=regressor.evaluate(test_data)
```

```python
pred_results.predictions.show()
```

```python
print(f"""
Linear Regression Results Report
==================================
RMSE:\t {pred_results.rootMeanSquaredError}
MSE:\t {pred_results.meanSquaredError}
R2:\t {pred_results.r2}
Adj R2:\t{pred_results.r2adj}
==================================
"""
)
```
